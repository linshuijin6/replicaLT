```python
import torch
import torch.nn as nn
from generative.networks.nets.diffusion_model_unet import DistributedDiffusionModelUNet
# 定义模型
model = DistributedDiffusionModelUNet(
    spatial_dims=3,
    in_channels=1,  # 1,
    out_channels=1,
    num_channels=(32, 64, 64, 128),
    attention_levels=(False, False, False, True),
    num_res_blocks=1,
    num_head_channels=(0, 0, 0, 128),
    with_conditioning=True,
    cross_attention_dim=512,
    use_flash_attention=True,
    device_ids=[0,1])

# 输入数据
x = torch.randn(1, 1, 160, 192, 160).to('cuda:0')  # CPU上即可
time = torch.randint(0, 1000, (1,)).to('cuda:0')
context = torch.randn(1, 2, 512).to('cuda:0')
out = model(x=x, timesteps=time, context=context)

loss = out.sum()
loss.backward()
```
